```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

%matplotlib inline
```

```python
image1 = mpimg.imread('robocam_1.jpg')
plt.imshow(image1)
```

```python
#Colors
red_ch = np.copy(image1)
red_ch[:,:,[1,2]]=0

green_ch = np.copy(image1)
green_ch[:,:,[0,2]]=0

blue_ch = np.copy(image1)
blue_ch[:,:,[0,1]]=0

fig = plt.figure(figsize=(12,3))
plt.subplot(131)
plt.imshow(red_ch)

plt.subplot(132)
plt.imshow(green_ch)

plt.subplot(133)
plt.imshow(blue_ch)
plt.show()
```

```python
def color_thresh(img, rgb_thresh=(0, 0, 0)):
    bin_img = np.zeros_like(image1[:,:,0])
    threshold = (image1[:,:,0] >= rgb_thresh[0]) & (image1[:,:,1] >= rgb_thresh[1]) & (image1[:,:,2] >= rgb_thresh[2])
    bin_img[threshold] = 1
    return bin_img
```

```python
thresholds = (160,160,160)
final_img = color_thresh(image1, rgb_thresh=thresholds)

fig2 = plt.figure(figsize=(21,7))
plt.subplot(121)
plt.imshow(image1)
plt.subplot(122)
plt.imshow(final_img,cmap='gray')
plt.show()
```
